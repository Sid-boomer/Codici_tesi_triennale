```python

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FuncFormatter

# Funzione per formattare i valori dell'asse y senza decimali
def integer_formatter(x, pos):
    if x.is_integer():
        return f'{int(x)}'  # Mostra solo numeri interi
    else:
        return f'{x}'
    
# Carica il file in un DataFrame, specificando i nomi delle colonne
data = pd.read_csv('Data_settembre\GL7-240924_132821.txt', sep='\t')

# Estrai le colonne di interesse
time = data['TIMESTAMP']  # La prima colonna come asse x
components = data.columns[1:8]  # Le sette colonne successive


print(data['TIMESTAMP'].min())
print(data['TIMESTAMP'].max())

# Crea il grafico
plt.figure(figsize=(12, 6))

# Ciclo sulle componenti, assegnando un colore diverso per ciascuna
for i, component in enumerate(components):
    plt.plot(time, data[component], label=component)

# Configura il grafico
plt.xlabel('Tempo (unit arbitrarie)')
plt.ylabel('Temperatura (K)')
plt.title('Ciclo di raffreddamento')
plt.yscale('log')
plt.legend()
plt.grid(True, which="both")

plt.gca().yaxis.set_major_formatter(FuncFormatter(integer_formatter))
#plt.gca().yaxis.get_major_formatter().set_scientific(False)
plt.yticks([ 0.1, 1, 10, 100, 1000])  # Imposta i tick sull'asse y a partire da 0.01
#plt.ylim(0.01, 1000)  # Imposta i limiti dell'asse y per partire da 0.01

plt.show()


# ZOOM
plt.figure(figsize=(12, 6))

for i, component in enumerate(components):
    plt.plot(time, data[component], label=component)


plt.xlabel('Tempo (unit arbitrarie)')
plt.ylabel('Temperatura (K)')
plt.title('')
plt.yscale('log')
plt.legend()
plt.grid(True, which="both")
plt.xlim(data['TIMESTAMP'].min(), data['TIMESTAMP'].max()-950000)
plt.gca().yaxis.set_major_formatter(FuncFormatter(integer_formatter))
#plt.gca().yaxis.get_major_formatter().set_scientific(False)
plt.yticks([ 0.1, 1, 10, 100])  # Imposta i tick sull'asse y a partire da 0.01
#plt.ylim(0.01, 1000)  # Imposta i limiti dell'asse y per partire da 0.01
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import glob
import os

# Funzione per formattare i valori dell'asse y senza decimali
def integer_formatter(x, pos):
    if x.is_integer():
        return f'{int(x)}'  # Mostra solo numeri interi
    else:
        return f'{x}'

# Percorso della cartella contenente i file
folder_path = 'Data_settembre'

# Cerca tutti i file .txt nella cartella
file_list = glob.glob(os.path.join(folder_path, '*.txt'))

# Itera su ciascun file trovato
for file_path in file_list:
    # Carica il file in un DataFrame
    data = pd.read_csv(file_path, sep='\t')

    # Estrai le colonne di interesse
    time = data['TIMESTAMP']  # La prima colonna come asse x
    components = data.columns[1:8]  # Le sette colonne successive

    # Crea il grafico per ogni file
    plt.figure(figsize=(12, 6))

    # Ciclo sulle componenti, assegnando un colore diverso per ciascuna
    for i, component in enumerate(components):
        plt.plot(time, data[component], label=component)

    # Configura il grafico
    plt.xlabel('Tempo (unit arbitrarie)')
    plt.ylabel('Temperatura (K)')
    plt.title(f'Ciclo di raffreddamento - {os.path.basename(file_path)}')
    plt.yscale('log')
    plt.legend()
    plt.grid(True, which="both")

    # Imposta i tick dell'asse y senza notazione scientifica
    plt.gca().yaxis.set_major_formatter(FuncFormatter(integer_formatter))
    plt.yticks([0.1, 1, 10, 100, 1000])  # Imposta i tick sull'asse y
    plt.ylim(0.1, 1000)  # Imposta i limiti dell'asse y
    
    plt.show()


    # Stampa i minimi e massimi per il timestamp del file attuale
    print(f"{os.path.basename(file_path)}: TIMESTAMP min = {data['TIMESTAMP'].min()}, TIMESTAMP max = {data['TIMESTAMP'].max()}")
```
